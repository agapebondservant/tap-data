### Analysis

The dataset amounts appear to be significantly right skewed, which is to be expected with an anomaly distribution. 75% of transactions fall under 77.165, and the middle 50% is between 22.0 and 77.165, but the maximum transaction is significantly higher than the 75th percentile. The median transaction is 22.0 while the mean is 88.3, also suggesting a significant right skew. Similarly, there is also a significant class imbalance, likely correlated to the amount imbalance.
