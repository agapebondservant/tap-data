```python
!pip install python-pdfbox transformers
import pdfbox
import logging
from transformers import AutoTokenizer, TFAutoModelForSeq2SeqLM, pipeline
import sys
```

```python
from google.colab import drive

drive.mount('/content/gdrive/', force_remount=True)
```

```python
def extract_text_from_pdf(file_path: str):
    extracted_file_path = file_path.replace('.pdf', '.txt')
    p = pdfbox.PDFBox()
    p.extract_text(file_path)
    with open(extracted_file_path, 'r') as handle:
        return handle.read()
```

```python
def summarize_pdf(file_path: str):
    content = extract_text_from_pdf(file_path)

    model = TFAutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
    tokenizer = AutoTokenizer.from_pretrained('facebook/bart-large-cnn')
    tokenizer.model_max_length = 1024 #sys.maxsize

    tokens_input = tokenizer.encode(f"summarize: {content}", return_tensors='tf', truncation=True)
    ids = model.generate(tokens_input, min_length=10000, max_length=20000)
    summary = tokenizer.decode(ids[0], skip_special_tokens=True)
    # summarizer = pipeline("summarization", max_length=500, model=model, tokenizer=tokenizer, truncation=True) https://drive.google.com/file/d/1XoyHW8HHKPva3EjP6jud7-HrAk109WpB/view?usp=sharing
    return summary
```

```python
summarize_pdf('/content/gdrive/MyDrive/covid1.pdf')
```
